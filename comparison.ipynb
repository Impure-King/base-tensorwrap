```python
# Creating a model:
class Model(eq.Module):
    def __init__(self):
        super().__init__()
        self.layer = eq.nn.Linear( 1,1, key =  1)
    def __call__(self, inputs, *args, **kwds):
        super().__call__(*args, **kwds)
        return self.layer(inputs)
```
